```python
from datetime import datetime
from meteostat import Point, Daily
```

```python
# Set time period
start = datetime(2015, 1, 1)
end = datetime(2023, 12, 31)
```

```python
# Create Point for Vila Militar, RJ - BR
dados = Point(-22.866251, -43.403784)
```

```python
# Get daily data (01/2015 - 12/2023)
data = Daily(dados, start, end)
data = data.fetch()
data.head(20)
```

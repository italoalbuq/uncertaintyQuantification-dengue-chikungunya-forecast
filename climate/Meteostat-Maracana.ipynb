```python
from datetime import datetime
from meteostat import Point, Daily
```

```python
# Set time period
start = datetime(2015, 1, 1)
end = datetime(2024, 6, 30)
```

```python
# Create Point for Maracan√£, RJ - BR
dados = Point(-22.913645, -43.225313)
```

```python
# Get daily data (01/2015 - 06/2024)
data = Daily(dados, start, end)
data = data.fetch()
data.head(20)
```
